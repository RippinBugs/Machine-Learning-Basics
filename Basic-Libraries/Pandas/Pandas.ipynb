#### Exercise 9
Find out the top five batsman with highest strike rates at the death overs using isin() method.

```python
delivery.head(3)
```

```python
"""
Death over balls dataframe
"""
death_over_mask = delivery['over'] > 15    #over between 16-20
#death_over_mask
death_over = delivery[death_over_mask]    #gives us all the deliveries at death overs.
```

```python
"""
calculating the total no of batsman who played at death overs.
"""
batsman = death_over.groupby('batsman').size() 
#batsman
```

```python
"""
Calculating the total no of batsman who played more than 200 balls at death overs.
"""
balls_mask = death_over.groupby('batsman')['batsman_runs'].count() > 200           #masking for min 200 balls.
#balls_mask
#batsman[balls_mask].shape      
#batsman[balls_mask]                                      #Gives us a series of 43 batsmen who played more than 200 balls.
batsman_list = batsman[balls_mask].index.tolist()   
```

```python
"""
Final dataframe for those 43 batsmen
"""
final_df = death_over[death_over['batsman'].isin(batsman_list)] #instead of filtering one value at a time we pass in a list
```

```python
balls = final_df.groupby('batsman').size()
runs = final_df.groupby('batsman')['batsman_runs'].sum()
```

```python
"""
Calculating Strike Rates for top 5 destructive batsmen at the death.
"""
strike_rate = (runs / balls) * 100
strike_rate.sort_values(ascending=False).round(2).head(5)
```

#### fillna() &nbsp; method

The fillna() method replaces the NULL values with a specified value. The fillna() method returns a new DataFrame object unless the inplace parameter is set to True , in that case the fillna() method does the replacing in the original DataFrame instead.

```python
titanic_data['Cabin'].fillna("Not specified",inplace = True)
```
