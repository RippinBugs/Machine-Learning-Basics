### Distribution plots 

```python
import seaborn as sns 
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#displot
sns.distplot(tips['total_bill'],kde= False,bins = 30)  #kernel distribution estimation
```

```python
#jointplot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
#pariplot
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
#rugplot
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.kdeplot(tips['total_bill'])
```

### matrix plot

```python
tips.head()
```

```python
fp = flights.pivot_table(index = 'month',columns='year',values='passengers')
```

###  Grid

```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
f = sns.PairGrid(iris)
f.map_diag(sns.distplot)
f.map_lower(sns.kdeplot)
f.map_upper(plt.scatter,color='red')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
g = sns.FacetGrid(tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g = sns.FacetGrid(tips,col='smoker',row='time')
g.map(plt.scatter,'total_bill','tip',color='pink')
```

### Seaborn exercises

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('whitegrid')
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.jointplot('fare','age',titanic)
```

```python
sns.distplot(titanic['fare'],bins=20,kde=False,color ='red')
```

```python
sns.boxplot(x='class',y='age',data = titanic)
```

```python
sns.swarmplot(x='class',y='age',data = titanic,palette='Set2')
```

```python
sns.countplot(x = 'sex',data = titanic)
```

```python
tc = titanic.corr()
```

```python
sns.heatmap(tc,cmap = 'coolwarm')
plt.title('titanic.corr()')
```

```python
g = sns.FacetGrid(data=titanic,col='sex')
g.map(plt.hist,'age')
```

```python
g=sns.FacetGrid(data = titanic,col='sex')
g.map(plt.hist,'age')
```
