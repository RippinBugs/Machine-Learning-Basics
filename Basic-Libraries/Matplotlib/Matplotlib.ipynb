### matplotlib exercises

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

plt.tight_layout
ax.plot(x,y)
ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax1.plot(x,y)
ax2.plot(x,y)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax1.plot(x,z)
ax2.plot(x,y)
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
ax2.set_title('zoom')
ax2.set_xlabel('x')
ax2.set_ylabel('z')
ax1.set_xlabel('x')
ax1.set_ylabel('z')
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
#ax[0]
#ax[1]

ax[0].plot(x,y,color = 'blue', ls='--',lw = 3)
ax[1].plot(x,z,color='red',ls = '-',lw=3)
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize = (8,2))
plt.tight_layout
#ax[0]
#ax[1]

ax[0].plot(x,y,color = 'blue', ls='-',lw = 3)
ax[1].plot(x,z,color='red',ls = '--',lw=3)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[1].set_xlabel('x')
ax[1].set_ylabel('z')
```
